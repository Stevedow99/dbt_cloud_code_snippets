### Connect to the dbt MCP Server

This cell connects to the deployed dbt-mcp app and lists all available tools. It uses `DatabricksMCPClient` which handles authentication to the Databricks-hosted MCP server via your workspace client.

**What to fill in:**
- **`mcp_server_url`** — The URL of your deployed app with `/mcp` appended. You can find this in the output of `databricks apps create` (the `url` field), or in the Databricks Apps UI. Depending on your cloud provider, it will look something like:
  - **Azure Databricks:** `https://custom-mcp-server-1234567890.4.azure.databricksapps.com/mcp`
  - **AWS Databricks:** `https://custom-mcp-server-1234567890.4.aws.databricksapps.com/mcp`

The `WorkspaceClient()` call with no arguments uses your current notebook authentication context automatically — no need to pass host, client ID, or client secret when running inside a Databricks notebook.

```python
import nest_asyncio
nest_asyncio.apply()

from databricks_mcp import DatabricksMCPClient
from databricks.sdk import WorkspaceClient

mcp_server_url = "https://<your-app-url>/mcp"

workspace_client = WorkspaceClient(
    host="https://<your-workspace-url>",
    client_id=dbutils.secrets.get(scope="dbt-mcp", key="oauth-client-id"),
    client_secret=dbutils.secrets.get(scope="dbt-mcp", key="oauth-client-secret")
)

mcp_client = DatabricksMCPClient(server_url=mcp_server_url, workspace_client=workspace_client)

tools = mcp_client.list_tools()
print(f"Found {len(tools)} tools:")
for tool in tools:
    print(f"  - {tool.name}")
```
